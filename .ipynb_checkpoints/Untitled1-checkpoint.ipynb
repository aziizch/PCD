```python
from huggingface_hub import HfApi, hf_hub_download
from tqdm import tqdm
import os

repo_id = "codefuse-ai/CodeFuse-CGE-Small"
repo_type = "model"
local_dir = r"C:\Users\clouduser\Desktop\CodeFuse-cgm\cge-small"

api = HfApi()
files = api.list_repo_files(repo_id=repo_id, repo_type=repo_type)

os.makedirs(local_dir, exist_ok=True)

for file in tqdm(files, desc="Downloading files"):
    hf_hub_download(
        repo_id=repo_id,
        repo_type=repo_type,
        filename=file,
        local_dir=local_dir,
        local_dir_use_symlinks=False,
        resume_download=True,
    )

print("Download finished")
```

```python
import json
import os

# 1. Mettez ici le chemin vers le DOSSIER du modèle CodeFuse-CGE-Large
# (Celui qui contient le fichier config.json)
model_path = r"C:\Users\clouduser\Desktop\CodeFuse-cgm\CodeFuse-CGM-main\cge-large" 
# NOTE : D'après votre erreur, le chemin semble être celui du cache ou de votre dossier téléchargé.
# Si vous avez téléchargé le modèle manuellement, mettez VOTRE chemin ici.
# Exemple : model_path = r"C:\Users\clouduser\Desktop\CodeFuse-cgm\CodeFuse-CGM-main"

# Essayons de trouver le bon chemin d'après votre erreur précédente
# Si vous utilisez le modèle téléchargé localement, décommentez et ajustez la ligne ci-dessous :
# model_path = r"C:\Users\clouduser\Desktop\CodeFuse-cgm\CodeFuse-CGM-main"

config_file = os.path.join(model_path, "config.json")

if os.path.exists(config_file):
    print(f"Fichier trouvé : {config_file}")
    
    # Lecture du fichier
    with open(config_file, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    # Vérification et correction
    if "rope_scaling" not in data:
        print("🔧 Le paramètre 'rope_scaling' est manquant. Correction en cours...")
        data["rope_scaling"] = None  # On ajoute le paramètre manquant
        
        # Sauvegarde
        with open(config_file, 'w', encoding='utf-8') as f:
            json.dump(data, f, indent=2)
        print("✅ Correction appliquée avec succès ! Réessayez votre script.")
    else:
        print("ℹ️ Le paramètre 'rope_scaling' existe déjà. Le problème vient d'ailleurs.")
else:
    print(f"❌ Erreur : Impossible de trouver le fichier config.json dans {model_path}")
    print("Vérifiez le chemin du dossier du modèle.")
```
